# JS 手撕题 · 第一天（基础）

今天的目标：用最基础的语法（变量、if、for、数组方法等）完成 5 道题。
建议做题方式：

1. 在这个单元格下面，新建一个 JavaScript 代码单元（code cell）。

2. 按顺序实现 `sum` / `maxInArray` / `unique` / `myForEach` / `myMap`。

3. 每写完一个函数，用 `console.log` 自己设计 2~3 个测试用例。

**题目 1：数组求和**

实现一个函数 `sum(arr)`，接收一个只包含数字的数组，返回所有元素的和。

示例：

- `sum([1, 2, 3])` 返回 `6`

- `sum([-1, 5, 10])` 返回 `14`

**题目 2：数组最大值**

实现一个函数 `maxInArray(arr)`，返回数组中的最大值（假设数组非空且都是数字）。

示例：

- `maxInArray([1, 7, 3])` 返回 `7`

- `maxInArray([-5, -2, -10])` 返回 `-2`

**题目 3：数组去重（基础版）**

实现一个函数 `unique(arr)`，返回一个新数组，去掉原数组中重复的元素（只考虑数字和字符串）。

示例：

- `unique([1, 1, 2, 3, 3])` 返回 `[1, 2, 3]`

- `unique(['a', 'b', 'a'])` 返回 `['a', 'b']`


**题目 4：实现 Array.prototype.myForEach**

给数组原型添加一个方法 `myForEach`，用法和原生 `forEach` 类似：

- 接收一个回调函数 `callback(element, index, array)`

- 对数组中的每一项依次调用回调

要求：**不能直接使用原生 `forEach` 实现**。



**题目 5：实现 Array.prototype.myMap**

给数组原型添加一个方法 `myMap`，用法和原生 `map` 类似：

- 接收一个回调函数 `callback(element, index, array)`

- 返回一个新数组，新数组的每一项为回调的返回值

要求：**不能直接使用原生 `map` 实现**。
